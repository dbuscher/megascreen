```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as pl
from numpy import pi
from MegaScreen import *
import functools

def Spectrum(f,f0,c=1,alpha=8./3.):
    c1=c*f0**(-alpha)
    a=c1*(1+alpha/2)
    b=c1*alpha/2/f0**2
    return np.where(f<f0,a-b*f**2,f**(-alpha))
    
f=np.linspace(0,3,200)
pl.plot(f,Spectrum(f,1.0))
```

```python
def Spectrum2(f,f0,f1=1e-3,c=1,alpha=8./6.):
    f2=(f0**2+f1**2)
    c1=c*f2**(-alpha)
    grad=2*alpha*f0*c1*f2**(-alpha-1)
    return np.where(f<f0,c1+2*grad*f0/pi*np.cos(pi*f/(2*f0)),(f**2+f1**2)**(-alpha))

f=np.linspace(0,3,200)
pl.plot(f,Spectrum2(f,1.0))
```
